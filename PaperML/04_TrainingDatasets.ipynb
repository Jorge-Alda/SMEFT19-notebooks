# 04: Preparing the Training datasets

## 2D Scans

The first set of data that we will include in the training datasets were used for [the likelihood plots](https://github.com/Jorge-Alda/SMEFT19/blob/master/jupyter/PaperML/02_plotfits.ipynb)

```python
import pandas as pd
import SMEFT19
```

```python
dbf = SMEFT19.ellipse.load('../data/ellipses/rotBII.yaml')
bf = dbf['bf']
```

```python
data_albl = pd.read_csv('../data/likelihood/likelihood_rotBII_albl.dat', sep='\t', header=None)
```

```python
data_albl.head()
```

```python
data_albl[9][4]
```

```python
dim_min = [-0.3, -0.08, -0.025, -0.15, 0]
dim_max = [0, 0.08, 0.025, 0.15, 3.0]
```

```python
xmin = dim_min[1]
xmax = dim_max[1]
ymin = dim_min[2]
ymax = dim_max[2]
xmargin = 0.02*(xmax-xmin)
ymargin = 0.02*(ymax-ymin)
dfsamplel = pd.DataFrame(columns=['C', 'al', 'bl', 'aq', 'bq', 'likelihood'])
likelihood = []
for num in range(50*50):
    ix = num % 50
    iy = num // 50
    x = (xmin-xmargin) + ix/50 * ((xmax+xmargin) - (xmin-xmargin))
    y = (ymin-ymargin) + iy/50 * ((ymax+ymargin) - (ymin-ymargin))
    dfsamplel = pd.concat([pd.DataFrame([[bf[0], x, y, bf[3], bf[4], data_albl[ix][iy]]], columns=dfsamplel.columns), dfsamplel], ignore_index=True)
    
```

```python
dfsamplel.describe()
```

```python
dfsamplel.to_csv('../data/samples/samplealbl.dat', sep='\t', index=False)
```

```python
data_aqbq = pd.read_csv('../data/likelihood/likelihood_rotBII_aqbq.dat', sep='\t', header=None)
data_aqbq.head()
```

```python
xmin = dim_min[3]
xmax = dim_max[3]
ymin = dim_min[4]
ymax = dim_max[4]
xmargin = 0.02*(xmax-xmin)
ymargin = 0.02*(ymax-ymin)
dfsampleq = pd.DataFrame(columns=['C', 'al', 'bl', 'aq', 'bq', 'likelihood'])
likelihood = []
for num in range(50*50):
    ix = num % 50
    iy = num // 50
    x = (xmin-xmargin) + ix/50 * ((xmax+xmargin) - (xmin-xmargin))
    y = (ymin-ymargin) + iy/50 * ((ymax+ymargin) - (ymin-ymargin))
    dfsampleq = pd.concat([pd.DataFrame([[bf[0], bf[1], bf[2], x, y, data_aqbq[ix][iy]]], columns=dfsampleq.columns), dfsampleq], ignore_index=True)
    
```

```python
dfsampleq.describe()
```

```python
dfsampleq.to_csv('../data/samples/sampleaqbq.dat', sep='\t', index=False)
```

```python
import pandas as pd
import imblearn
```

```python
dataset = ['../data/samples/randompoints_III.dat', '../data/samples/sample_rotBIII_C1C3.dat', '../data/samples/sample_rotBIII_C1bq.dat', '../data/samples/sample_rotBIII_C3bq.dat']
dfs = [pd.read_csv(f, sep='\t', names=['C1', 'C3', 'bq', 'logL'], header=0) for i, f in enumerate(dataset)]
df = pd.concat(dfs, ignore_index=True)
df = df.loc[df['logL'] > 10.0]
```

```python
df['class'] = df['logL'] > 18.0
```

```python
df
```

```python
ros = imblearn.over_sampling.RandomOverSampler(sampling_strategy=1.0)
```

```python
X = df[['C1', 'C3', 'bq', 'logL']]
Y = df['class']
```

```python
X_over, Y_over = ros.fit_resample(X, Y)
```

```python
X_over.loc[X_over['logL']>18]
```

```python
X_over.to_csv('../data/samples/training_III.dat', sep='\t', header=False, index=False)
```
