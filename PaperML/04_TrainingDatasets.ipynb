# 04: Preparing the Training datasets

## 2D Scans

The first set of data that we will include in the training datasets were used for [the likelihood plots](https://github.com/Jorge-Alda/SMEFT19/blob/master/jupyter/PaperML/02_plotfits.ipynb)

```python
import pandas as pd
import SMEFT19
```

```python
dbf = SMEFT19.ellipse.load('../../ellipses/rotBII.yaml')
bf = dbf['bf']
```

```python
data_albl = pd.read_csv('../../likelihood/likelihood_rotBII_l.dat', sep='\t')
```

```python
data_albl.head()
```

```python
dfsamplel = pd.DataFrame()
```

```python
numpoints = len(data_albl)
```

```python
dfsamplel['C'] = [bf[0]]*numpoints
dfsamplel['al'] = data_albl['x']
dfsamplel['bl'] = data_albl['y']
dfsamplel['aq'] = [bf[3]]*numpoints
dfsamplel['bq'] = [bf[4]]*numpoints
dfsamplel['likelihood'] = data_albl['global']
```

```python
dfsamplel.head()
```

```python
dfsamplel.to_csv('../../samples/samplealbl.dat', sep='\t', index=False)
```

```python
data_aqbq = pd.read_csv('../../likelihood/likelihood_rotBII_q.dat', sep='\t')
data_aqbq.head()
```

```python
dfsampleq = pd.DataFrame()
numpoints = len(data_aqbq)
dfsampleq['C'] = [bf[0]]*numpoints
dfsampleq['al'] = [bf[1]]*numpoints
dfsampleq['bl'] = [bf[2]]*numpoints
dfsampleq['aq'] = data_aqbq['x']
dfsampleq['bq'] = data_aqbq['y']
dfsampleq['likelihood'] = data_aqbq['global']
```

```python
dfsampleq.head()
```

```python
dfsampleq.to_csv('../../samples/sampleaqbq.dat', sep='\t', index=False)
```

## Random points

```python
from tqdm.notebook import tqdm
from parscanning import RandomScan
```

```python
RS = RandomScan(SMEFT19.likelihood_global, [-0.22, -0.15, -0.03, -0.12, 0],
                [-0.04, 0.15, 0.03, 0.075, 3.5], 10)
```

```python
for i in tqdm(range(111)):
    RS.run_mp(4, SMEFT19.scenarios.rotBII)
    RS.write('../../samples/randompoints.dat', 'at')
    RS.clear()
```
